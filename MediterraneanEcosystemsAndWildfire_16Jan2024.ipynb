```python
import pandas as pd
import geopandas as gpd
import contextily as cx
```

### Use the RESOLVE dataset to map Mediterranean biomes
https://ecoregions.appspot.com/

```python
MyWorkingDir = "Data\\"
```

```python
med_biomes_f = f'{MyWorkingDir}REVEAL_Med_Biomes.shp'
med_biomes = gpd.read_file(med_biomes_f)
med_biomes = med_biomes.to_crs(epsg=4326)
```

```python
ax = med_biomes.plot(figsize=(10, 10),alpha=0.5,color=med_biomes.COLOR)
cx.add_basemap(ax, source=cx.providers.CartoDB.Positron,crs=med_biomes.crs,zoom=4)
ax.grid()
ax.set_title("Mediterranean Biomes (RESOLVE)")
```

### Subset data into regions:
1. North America: United States and Mexico
1. South America: Chile
1. Mediterranean basin
1. Southern Africa: 
1. Australia


```python
# North America
med_biomes_na = med_biomes.cx[-135:-25, 45:25]
ax = med_biomes_na.plot(color=med_biomes_na.COLOR,alpha=0.5)
cx.add_basemap(ax, source=cx.providers.CartoDB.Positron,crs=med_biomes_na.crs,zoom=6)
```

```python
# South America
med_biomes_sa = med_biomes.cx[-135:-25, -25:-45]
ax = med_biomes_sa.plot(color=med_biomes_sa.COLOR,alpha=0.5)
cx.add_basemap(ax, source=cx.providers.CartoDB.Positron,crs=med_biomes_sa.crs,zoom=6)
```

```python
minx, miny, maxx, maxy = med_biomes_na.geometry.total_bounds
minx, miny, maxx, maxy
```

```python
# South Africa
med_biomes_saf = med_biomes.cx[-25:50, -25:-45]
ax = med_biomes_saf.plot(color=med_biomes_saf.COLOR,alpha=0.5)
cx.add_basemap(ax, source=cx.providers.CartoDB.Positron,crs=med_biomes_saf.crs,zoom=6)
```

```python
# Mediterranean basin
med_biomes_mb = med_biomes.cx[-25:50, 45:25]
ax = med_biomes_mb.plot(color=med_biomes_mb.COLOR,alpha=0.5)
cx.add_basemap(ax, source=cx.providers.CartoDB.Positron,crs=med_biomes_mb.crs,zoom=6)
```

```python
# Australia
med_biomes_aus = med_biomes.cx[100:180, -25:-45]
ax = med_biomes_aus.plot(color=med_biomes_aus.COLOR,alpha=0.5)
cx.add_basemap(ax, source=cx.providers.CartoDB.Positron,crs=med_biomes_aus.crs,zoom=6)
```

```python
def WriteRegionShapefile(gdf,region,MyGoogleDrive):
    f = f'{MyWorkingDir}REVEAL_Med_Biomes_{region}.shp'
    gdf.to_file(f)
    
```

```python
WriteRegionShapefile(med_biomes_na,"na",MyGoogleDrive)
WriteRegionShapefile(med_biomes_sa,"sa",MyGoogleDrive)
WriteRegionShapefile(med_biomes_saf,"saf",MyGoogleDrive)
WriteRegionShapefile(med_biomes_mb,"mb",MyGoogleDrive)
WriteRegionShapefile(med_biomes_aus,"aus",MyGoogleDrive)
```
